```python
documents = {}

with open("documents.txt", "r") as file:
    for line in file:
        doc_id, content = line.strip().split("|", 1)
        documents[int(doc_id)] = content

for doc_id, content in documents.items():
    print(f"Document {doc_id}: {content}")
```

```python
def preprocess_text(text):
    return text.lower().split()

print("Preprocessed Documents:")

for doc_id, content in documents.items():
    tokens = preprocess_text(content)
    print(f"Document {doc_id}: {tokens}")
```

**2. Inverted Index:**

```python
inverted_index = {}

for doc_id, content in documents.items():
    words = preprocess_text(content)
    for word in words:
        if word not in inverted_index:
            inverted_index[word] = set()
        inverted_index[word].add(doc_id)

print("Inverted Index built successfully")
print("Total unique terms:", len(inverted_index))
```

```python
print("Displaying Inverted Index:")

for term in sorted(inverted_index):
    print(f"{term} : {sorted(inverted_index[term])}")
```

**3. Boolean Query Processing:**

```python
print("Boolean Query Processor initialized successfully")

def process_boolean_query(query, inverted_index, all_docs):
    tokens = query.lower().split()
    result = None
    operator = None

    for token in tokens:
        if token in ["and", "or", "not"]:
            operator = token
        else:
            docs = inverted_index.get(token, set())

            if operator == "not":
                docs = all_docs - docs
                operator = None

            if result is None:
                result = docs
            else:
                if operator == "and":
                    result = result & docs
                elif operator == "or":
                    result = result | docs
                operator = None

    return result
```

```python
query = input("Enter a Boolean query: ")

all_docs = set(documents.keys())

result = process_boolean_query(query, inverted_index, all_docs)

print("\nQuery entered:", query)
print("Matching Document IDs:", result)
```
